```python
import os

RAW_VISDRONE = f'{DRIVE_BASE}/datasets/raw/visdrone'

VIS_IMAGES = f'{DRIVE_BASE}/datasets/processed/visdrone/all/images'
VIS_LABELS = f'{DRIVE_BASE}/datasets/processed/visdrone/all/labels'

os.makedirs(VIS_IMAGES, exist_ok=True)
os.makedirs(VIS_LABELS, exist_ok=True)

print(os.path.exists(RAW_VISDRONE))
print(VIS_IMAGES)
print(VIS_LABELS)
```

```python
from PIL import Image
from pathlib import Path
import shutil, os

def visdrone2yolo_person_only(src_dir, dst_images_dir, dst_labels_dir):
    ann_dir = Path(src_dir) / 'annotations'
    img_dir = Path(src_dir) / 'images'

    os.makedirs(dst_images_dir, exist_ok=True)
    os.makedirs(dst_labels_dir, exist_ok=True)

    converted = 0

    for f in ann_dir.glob('*.txt'):
        img_path = img_dir / f.with_suffix('.jpg').name

        if not img_path.exists():
            continue

        img_size = Image.open(img_path).size
        dw, dh = 1.0 / img_size[0], 1.0 / img_size[1]
        lines = []

        with open(f, encoding='utf-8') as file:
            for row in [x.split(',') for x in file.read().strip().splitlines()]:
                if len(row) < 6:
                    continue
                if row[4] == '0':        # skip ignored regions
                    continue
                cls = int(row[5]) - 1    # pedestrian(1)->0, people(2)->1
                if cls not in [0, 1]:    # keep only pedestrian and people
                    continue
                x, y, w, h = map(int, row[:4])
                xc = (x + w / 2) * dw
                yc = (y + h / 2) * dh
                wn = w * dw
                hn = h * dh
                lines.append(f'0 {xc:.6f} {yc:.6f} {wn:.6f} {hn:.6f}\n')

        if lines:
            shutil.copy2(img_path, Path(dst_images_dir) / img_path.name)
            (Path(dst_labels_dir) / f.name).write_text(''.join(lines), encoding='utf-8')
            converted += 1

    return converted

# Run for all three splits
total = 0
for folder in ['VisDrone2019-DET-train', 'VisDrone2019-DET-val', 'VisDrone2019-DET-test-dev']:
    src = f'{RAW_VISDRONE}/{folder}'
    if not os.path.exists(src):
        print(f'NOT FOUND: {src}')
        continue
    print(f'Converting {folder}...')
    n = visdrone2yolo_person_only(src, VIS_IMAGES, VIS_LABELS)
    print(f'  → {n} images with person annotations')
    total += n

print(f'\nTotal: {total} images copied to pool')
```

```python
import cv2
import os

def check_corrupted_images(images_dir):
    corrupted = []
    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img_path = os.path.join(images_dir, filename)
        img = cv2.imread(img_path)
        if img is None:
            corrupted.append(filename)

    print(f'Total images checked: {len(os.listdir(images_dir))}')
    print(f'Corrupted images found: {len(corrupted)}')
    for f in corrupted:
        print(f'  {f}')
    return corrupted

def check_corrupted_labels(labels_dir):
    empty = []
    malformed = []

    for filename in os.listdir(labels_dir):
        if not filename.endswith('.txt'):
            continue
        lbl_path = os.path.join(labels_dir, filename)

        with open(lbl_path) as f:
            lines = f.readlines()

        if len(lines) == 0:
            empty.append(filename)
            continue

        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                malformed.append(filename)
                break
            try:
                float(parts[1])
                float(parts[2])
                float(parts[3])
                float(parts[4])
            except ValueError:
                malformed.append(filename)
                break

    print(f'Total labels checked: {len(os.listdir(labels_dir))}')
    print(f'Empty label files: {len(empty)}')
    print(f'Malformed label files: {len(malformed)}')
    return empty, malformed
```

```python
def check_invalid_boxes(labels_dir):
    invalid = []

    for filename in os.listdir(labels_dir):
        if not filename.endswith('.txt'):
            continue
        lbl_path = os.path.join(labels_dir, filename)

        with open(lbl_path) as f:
            lines = f.readlines()

        for i, line in enumerate(lines):
            parts = line.strip().split()
            if len(parts) != 5:
                continue

            cls = int(parts[0])
            xc = float(parts[1])
            yc = float(parts[2])
            w  = float(parts[3])
            h  = float(parts[4])

            reasons = []

            if w <= 0 or h <= 0:
                reasons.append('zero or negative dimension')
            if xc - w/2 < 0:
                reasons.append('box exceeds left boundary')
            if xc + w/2 > 1:
                reasons.append('box exceeds right boundary')
            if yc - h/2 < 0:
                reasons.append('box exceeds top boundary')
            if yc + h/2 > 1:
                reasons.append('box exceeds bottom boundary')
            if w * h > 0.95:
                reasons.append('box covers >95% of image')

            if reasons:
                invalid.append((filename, i+1, reasons))

    print(f'Files with invalid boxes: {len(set(f for f,_,_ in invalid))}')
    print(f'Total invalid box instances: {len(invalid)}')
    for filename, line_num, reasons in invalid[:10]:
        print(f'  {filename} line {line_num}: {reasons}')
    return invalid
```

```python
def check_grayscale_images(images_dir):
    grayscale = []

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img_path = os.path.join(images_dir, filename)
        img = cv2.imread(img_path)
        if img is None:
            continue

        if len(img.shape) == 2:
            grayscale.append(filename)
        elif img.shape[2] != 3:
            grayscale.append(filename)

    print(f'Grayscale or non-RGB images found: {len(grayscale)}')
    for f in grayscale:
        print(f'  {f}')
    return grayscale
```

```python
from collections import Counter

def check_resolutions(images_dir):
    resolutions = Counter()

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img_path = os.path.join(images_dir, filename)
        img = cv2.imread(img_path)
        if img is None:
            continue
        h, w = img.shape[:2]
        resolutions[(w, h)] += 1

    print(f'Resolution distribution:')
    for res, count in resolutions.most_common(10):
        print(f'  {res[0]}x{res[1]}: {count} images')
    return resolutions
```

```python
import hashlib

def check_duplicates(images_dir):
    hashes = {}
    filename_duplicates = []
    content_duplicates = []

    seen_names = set()

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        # filename duplicate
        if filename in seen_names:
            filename_duplicates.append(filename)
        seen_names.add(filename)

        # content duplicate
        img_path = os.path.join(images_dir, filename)
        with open(img_path, 'rb') as f:
            file_hash = hashlib.md5(f.read()).hexdigest()

        if file_hash in hashes:
            content_duplicates.append((filename, hashes[file_hash]))
        else:
            hashes[file_hash] = filename

    print(f'Filename duplicates: {len(filename_duplicates)}')
    print(f'Content duplicates: {len(content_duplicates)}')
    for dup, original in content_duplicates[:10]:
        print(f'  {dup} is duplicate of {original}')
    return filename_duplicates, content_duplicates
```

```python
def check_label_image_consistency(images_dir, labels_dir):
    imgs = set(os.path.splitext(f)[0] for f in os.listdir(images_dir)
               if f.lower().endswith(('.jpg', '.jpeg', '.png')))
    lbls = set(os.path.splitext(f)[0] for f in os.listdir(labels_dir)
               if f.endswith('.txt'))

    imgs_no_label = imgs - lbls
    lbls_no_image = lbls - imgs

    print(f'Total images: {len(imgs)}')
    print(f'Total labels: {len(lbls)}')
    print(f'Images without label: {len(imgs_no_label)}')
    print(f'Labels without image: {len(lbls_no_image)}')

    if imgs_no_label:
        print('Samples — images without label:')
        for f in list(imgs_no_label)[:5]:
            print(f'  {f}')
    if lbls_no_image:
        print('Samples — labels without image:')
        for f in list(lbls_no_image)[:5]:
            print(f'  {f}')

    return imgs_no_label, lbls_no_image
```

```python
def check_class_consistency(labels_dir):
    wrong_class = []

    for filename in os.listdir(labels_dir):
        if not filename.endswith('.txt'):
            continue
        with open(os.path.join(labels_dir, filename)) as f:
            for i, line in enumerate(f):
                parts = line.strip().split()
                if len(parts) == 5 and parts[0] != '0':
                    wrong_class.append((filename, i+1, parts[0]))

    print(f'Total label files scanned: {len(os.listdir(labels_dir))}')
    print(f'Lines with wrong class ID: {len(wrong_class)}')
    if wrong_class:
        print('Samples:')
        for f, line, cls in wrong_class[:10]:
            print(f'  {f} line {line}: class {cls}')
    else:
        print('All lines are class 0. ✅')

    return wrong_class
```

```python
def clip_invalid_boxes(labels_dir):
    clipped_files = 0
    clipped_boxes = 0

    for filename in os.listdir(labels_dir):
        if not filename.endswith('.txt'):
            continue
        lbl_path = os.path.join(labels_dir, filename)

        with open(lbl_path) as f:
            lines = f.readlines()

        new_lines = []
        file_changed = False

        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                new_lines.append(line)
                continue

            cls = parts[0]
            xc = float(parts[1])
            yc = float(parts[2])
            w  = float(parts[3])
            h  = float(parts[4])

            x_min = max(0.0, xc - w/2)
            x_max = min(1.0, xc + w/2)
            y_min = max(0.0, yc - h/2)
            y_max = min(1.0, yc + h/2)

            new_w = x_max - x_min
            new_h = y_max - y_min
            new_xc = x_min + new_w/2
            new_yc = y_min + new_h/2

            if new_w <= 0 or new_h <= 0:
                file_changed = True
                clipped_boxes += 1
                continue

            # ✅ THIS IS THE FIX — compare floats not strings
            was_clipped = (
                abs(new_xc - xc) > 1e-9 or
                abs(new_yc - yc) > 1e-9 or
                abs(new_w  - w)  > 1e-9 or
                abs(new_h  - h)  > 1e-9
            )

            new_line = f'{cls} {new_xc:.6f} {new_yc:.6f} {new_w:.6f} {new_h:.6f}\n'

            if was_clipped:
                file_changed = True
                clipped_boxes += 1

            new_lines.append(new_line)

        if file_changed:
            with open(lbl_path, 'w') as f:
                f.writelines(new_lines)
            clipped_files += 1

    print(f'Files modified: {clipped_files}')
    print(f'Boxes clipped or removed: {clipped_boxes}')
```

```python
import hashlib

def remove_content_duplicates(images_dir, labels_dir):
    hashes = {}
    removed = 0

    for filename in sorted(os.listdir(images_dir)):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        img_path = os.path.join(images_dir, filename)
        with open(img_path, 'rb') as f:
            file_hash = hashlib.md5(f.read()).hexdigest()

        if file_hash in hashes:
            # this is a duplicate — remove it
            stem = os.path.splitext(filename)[0]
            lbl_path = os.path.join(labels_dir, f'{stem}.txt')

            os.remove(img_path)
            if os.path.exists(lbl_path):
                os.remove(lbl_path)

            print(f'Removed duplicate: {filename} (original: {hashes[file_hash]})')
            removed += 1
        else:
            hashes[file_hash] = filename

    print(f'Total duplicates removed: {removed}')
```

```python
def check_brightness_outliers(images_dir, dataset_name,
                               low_threshold=5, high_threshold=240,
                               n_samples=5):
    too_dark = []
    too_bright = []

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img = cv2.imread(os.path.join(images_dir, filename))
        if img is None:
            continue
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        brightness = np.mean(hsv[:,:,2])

        if brightness < low_threshold:
            too_dark.append((filename, brightness))
        elif brightness > high_threshold:
            too_bright.append((filename, brightness))

    print(f'\nBrightness Outliers — {dataset_name}')
    print(f'Nearly black (brightness < {low_threshold}): {len(too_dark)}')
    print(f'Nearly white (brightness > {high_threshold}): {len(too_bright)}')

    # visual spot check
    samples = too_dark[:n_samples] + too_bright[:n_samples]
    if samples:
        fig, axes = plt.subplots(1, len(samples), figsize=(4*len(samples), 4))
        if len(samples) == 1:
            axes = [axes]
        for ax, (filename, brightness) in zip(axes, samples):
            img = cv2.imread(os.path.join(images_dir, filename))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            ax.imshow(img)
            ax.set_title(f'brightness={brightness:.1f}', fontsize=8)
            ax.axis('off')
        plt.suptitle(f'Brightness Outliers — {dataset_name}')
        plt.tight_layout()
        plt.show()
    else:
        print('No brightness outliers found.')

    return too_dark, too_bright
```

```python
def check_scale_outliers(images_dir, labels_dir, dataset_name,
                          tiny_threshold=100, giant_threshold=0.5):
    """
    tiny_threshold: box area in pixels² below which a box is flagged as tiny
    giant_threshold: box area as fraction of image area above which flagged as giant
    """
    all_tiny_images = []
    all_giant_images = []

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img = cv2.imread(os.path.join(images_dir, filename))
        if img is None:
            continue
        h, w = img.shape[:2]
        image_area = h * w

        stem = os.path.splitext(filename)[0]
        lbl_path = os.path.join(labels_dir, f'{stem}.txt')
        if not os.path.exists(lbl_path):
            continue

        with open(lbl_path) as f:
            lines = [l.strip().split() for l in f if l.strip()]

        if not lines:
            continue

        areas = []
        for parts in lines:
            if len(parts) != 5:
                continue
            pw = float(parts[3]) * w
            ph = float(parts[4]) * h
            areas.append(pw * ph)

        if not areas:
            continue

        # flag image if ALL boxes are tiny
        if all(a < tiny_threshold for a in areas):
            all_tiny_images.append((filename, np.mean(areas)))

        # flag image if ANY box covers more than giant_threshold of image
        if any(a / image_area > giant_threshold for a in areas):
            all_giant_images.append((filename, max(areas)/image_area))

    print(f'\nScale Outliers — {dataset_name}')
    print(f'Images where ALL boxes are tiny (< {tiny_threshold}px²): '
          f'{len(all_tiny_images)}')
    print(f'Images with a giant box (> {giant_threshold*100:.0f}% of image): '
          f'{len(all_giant_images)}')

    # visual spot check — show 3 samples of each
    def show_samples(sample_list, title, n=3):
        samples = sample_list[:n]
        if not samples:
            return
        fig, axes = plt.subplots(1, len(samples),
                                  figsize=(5*len(samples), 4))
        if len(samples) == 1:
            axes = [axes]
        for ax, (filename, val) in zip(axes, samples):
            img = cv2.imread(os.path.join(images_dir, filename))
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # draw boxes
            stem = os.path.splitext(filename)[0]
            lbl_path = os.path.join(labels_dir, f'{stem}.txt')
            ih, iw = img.shape[:2]
            with open(lbl_path) as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) != 5:
                        continue
                    xc = float(parts[1]) * iw
                    yc = float(parts[2]) * ih
                    bw = float(parts[3]) * iw
                    bh = float(parts[4]) * ih
                    x1 = int(xc - bw/2)
                    y1 = int(yc - bh/2)
                    x2 = int(xc + bw/2)
                    y2 = int(yc + bh/2)
                    cv2.rectangle(img_rgb, (x1,y1), (x2,y2), (0,255,0), 2)

            ax.imshow(img_rgb)
            ax.set_title(f'{filename[:20]}\nval={val:.2f}', fontsize=7)
            ax.axis('off')

        plt.suptitle(title)
        plt.tight_layout()
        plt.show()

    show_samples(all_tiny_images,
                 f'All-Tiny Box Images — {dataset_name}')
    show_samples(all_giant_images,
                 f'Giant Box Images — {dataset_name}')

    return all_tiny_images, all_giant_images
```

```python
def compute_illumination(images_dir, dataset_name):
    brightness_values = []
    contrast_values = []

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img = cv2.imread(os.path.join(images_dir, filename))
        if img is None:
            continue
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        v = hsv[:,:,2]
        brightness_values.append(float(np.mean(v)))
        contrast_values.append(float(np.std(v)))

    print(f'\nIllumination Analysis — {dataset_name}')
    print(f'Total images: {len(brightness_values)}')
    print(f'Mean brightness: {np.mean(brightness_values):.2f}')
    print(f'Std brightness: {np.std(brightness_values):.2f}')
    print(f'Mean contrast: {np.mean(contrast_values):.2f}')
    print(f'Std contrast: {np.std(contrast_values):.2f}')

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].hist(brightness_values, bins=30, color='steelblue', edgecolor='black')
    axes[0].set_title(f'Brightness Distribution — {dataset_name}')
    axes[0].set_xlabel('Brightness (V channel mean)')
    axes[0].set_ylabel('Number of images')
    axes[1].hist(contrast_values, bins=30, color='coral', edgecolor='black')
    axes[1].set_title(f'Contrast Distribution — {dataset_name}')
    axes[1].set_xlabel('Contrast (V channel std)')
    axes[1].set_ylabel('Number of images')
    plt.tight_layout()
    plt.savefig(f'{dataset_name}_illumination.png', dpi=150, bbox_inches='tight')
    plt.show()

    return brightness_values, contrast_values
```

```python
def compute_illumination_categories(brightness_values, dataset_name):
    import numpy as np

    p20 = np.percentile(brightness_values, 20)
    p40 = np.percentile(brightness_values, 40)
    p60 = np.percentile(brightness_values, 60)
    p80 = np.percentile(brightness_values, 80)

    categories = {
        f'Very Dark (0-{p20:.1f})':       0,
        f'Dark ({p20:.1f}-{p40:.1f})':    0,
        f'Moderate ({p40:.1f}-{p60:.1f})':0,
        f'Bright ({p60:.1f}-{p80:.1f})':  0,
        f'Very Bright ({p80:.1f}+)':      0,
    }
    keys = list(categories.keys())

    for b in brightness_values:
        if b < p20:
            categories[keys[0]] += 1
        elif b < p40:
            categories[keys[1]] += 1
        elif b < p60:
            categories[keys[2]] += 1
        elif b < p80:
            categories[keys[3]] += 1
        else:
            categories[keys[4]] += 1

    total = len(brightness_values)
    print(f'\nIllumination Categories — {dataset_name}')
    for cat, count in categories.items():
        print(f'  {cat}: {count} images ({count/total*100:.1f}%)')

    plt.figure(figsize=(10, 5))
    plt.bar(categories.keys(), categories.values(),
            color=['#1a1a2e','#16213e','#0f3460','#e94560','#f5a623'],
            edgecolor='black')
    plt.title(f'Illumination Category Distribution — {dataset_name}')
    plt.xlabel('Illumination Category')
    plt.ylabel('Number of images')
    plt.xticks(rotation=15)
    plt.tight_layout()
    plt.savefig(f'{dataset_name}_illumination_categories.png', dpi=150, bbox_inches='tight')
    plt.show()

    return categories, (p20, p40, p60, p80)
```

```python
def compute_bbox_scale(images_dir, labels_dir, dataset_name):
    small, medium, large = 0, 0, 0
    box_areas = []

    for filename in os.listdir(images_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img = cv2.imread(os.path.join(images_dir, filename))
        if img is None:
            continue
        h, w = img.shape[:2]
        stem = os.path.splitext(filename)[0]
        lbl_path = os.path.join(labels_dir, f'{stem}.txt')
        if not os.path.exists(lbl_path):
            continue
        with open(lbl_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue
                pw = float(parts[3]) * w
                ph = float(parts[4]) * h
                area = pw * ph
                box_areas.append(area)
                if area < 1024:
                    small += 1
                elif area < 9216:
                    medium += 1
                else:
                    large += 1

    total = small + medium + large
    print(f'\nBounding Box Scale — {dataset_name}')
    print(f'Total boxes: {total}')
    print(f'Small  (< 32x32):     {small}  ({small/total*100:.1f}%)')
    print(f'Medium (32x32-96x96): {medium} ({medium/total*100:.1f}%)')
    print(f'Large  (> 96x96):     {large}  ({large/total*100:.1f}%)')

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].bar(['Small', 'Medium', 'Large'], [small, medium, large],
                color=['#e74c3c','#3498db','#2ecc71'], edgecolor='black')
    axes[0].set_title(f'Box Scale Distribution — {dataset_name}')
    axes[0].set_xlabel('Scale Category')
    axes[0].set_ylabel('Number of boxes')
    axes[1].hist(box_areas, bins=100, color='steelblue', edgecolor='black')
    axes[1].axvline(x=1024, color='red', linestyle='--', linewidth=2,
                    label='Small threshold (32x32)')
    axes[1].axvline(x=9216, color='orange', linestyle='--', linewidth=2,
                    label='Medium threshold (96x96)')
    axes[1].set_title(f'Box Area Distribution — {dataset_name}')
    axes[1].set_xlabel('Box Area (pixels²)')
    axes[1].set_ylabel('Number of boxes')
    axes[1].set_xlim(0, 50000)
    axes[1].legend()
    plt.tight_layout()
    plt.savefig(f'{dataset_name}_bbox_scale.png', dpi=150, bbox_inches='tight')
    plt.show()

    return box_areas, small, medium, large
```

```python
def compute_bbox_aspect_ratio(labels_dir, dataset_name):
    ratios = []
    for filename in os.listdir(labels_dir):
        if not filename.endswith('.txt'):
            continue
        with open(os.path.join(labels_dir, filename)) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue
                w = float(parts[3])
                h = float(parts[4])
                if h > 0:
                    ratios.append(w / h)

    print(f'\nBounding Box Aspect Ratio — {dataset_name}')
    print(f'Total boxes: {len(ratios)}')
    print(f'Mean w/h ratio: {np.mean(ratios):.3f}')
    print(f'Median w/h ratio: {np.median(ratios):.3f}')
    print(f'Tall (w/h < 0.5):    {sum(1 for r in ratios if r < 0.5)} ({sum(1 for r in ratios if r < 0.5)/len(ratios)*100:.1f}%)')
    print(f'Square (0.5-1.5):    {sum(1 for r in ratios if 0.5 <= r <= 1.5)} ({sum(1 for r in ratios if 0.5 <= r <= 1.5)/len(ratios)*100:.1f}%)')
    print(f'Wide (w/h > 1.5):    {sum(1 for r in ratios if r > 1.5)} ({sum(1 for r in ratios if r > 1.5)/len(ratios)*100:.1f}%)')

    plt.figure(figsize=(10, 4))
    plt.hist(ratios, bins=50, color='mediumpurple', edgecolor='black')
    plt.axvline(x=1.0, color='red', linestyle='--', linewidth=2, label='Square (w=h)')
    plt.title(f'Bounding Box Aspect Ratio (w/h) — {dataset_name}')
    plt.xlabel('Width / Height ratio')
    plt.ylabel('Number of boxes')
    plt.xlim(0, 3)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'{dataset_name}_bbox_aspect_ratio.png', dpi=150, bbox_inches='tight')
    plt.show()

    return ratios
```

```python
def compute_spatial_heatmap(labels_dir, dataset_name, grid_size=20):
    heatmap = np.zeros((grid_size, grid_size))
    for filename in os.listdir(labels_dir):
        if not filename.endswith('.txt'):
            continue
        with open(os.path.join(labels_dir, filename)) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue
                xc = float(parts[1])
                yc = float(parts[2])
                col = min(int(xc * grid_size), grid_size - 1)
                row = min(int(yc * grid_size), grid_size - 1)
                heatmap[row, col] += 1

    plt.figure(figsize=(8, 6))
    plt.imshow(heatmap, cmap='hot', interpolation='nearest')
    plt.colorbar(label='Number of annotations')
    plt.title(f'Bounding Box Center Spatial Distribution — {dataset_name}')
    plt.xlabel('X position (normalized)')
    plt.ylabel('Y position (normalized)')
    plt.tight_layout()
    plt.savefig(f'{dataset_name}_spatial_heatmap.png', dpi=150, bbox_inches='tight')
    plt.show()

    return heatmap
```

```python
def compare_illumination_overlap(brightness_dict):
    colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']
    plt.figure(figsize=(12, 5))
    for i, (name, values) in enumerate(brightness_dict.items()):
        plt.hist(values, bins=30, alpha=0.5,
                 color=colors[i], label=name, edgecolor='none')
    plt.title('Illumination Distribution Overlap Across Datasets')
    plt.xlabel('Brightness (V channel mean)')
    plt.ylabel('Number of images')
    plt.legend()
    plt.tight_layout()
    plt.savefig('illumination_overlap.png', dpi=150, bbox_inches='tight')
    plt.show()

    print('\nIllumination Range Per Dataset:')
    for name, values in brightness_dict.items():
        print(f'  {name}: min={min(values):.1f}, max={max(values):.1f}, mean={np.mean(values):.1f}')
```

```python
def print_eda_summary(results):
    print('\n' + '='*70)
    print('DATASET EDA SUMMARY')
    print('='*70)
    print(f"{'Metric':<30}", end='')
    for name in results:
        print(f"{name:<20}", end='')
    print()
    print('-'*70)

    metrics = [
        ('Total images',      'images'),
        ('Total annotations', 'annotations'),
        ('Mean brightness',   'mean_brightness'),
        ('Mean contrast',     'mean_contrast'),
        ('Mean boxes/image',  'mean_boxes'),
        ('Small boxes %',     'small_pct'),
        ('Medium boxes %',    'medium_pct'),
        ('Large boxes %',     'large_pct'),
    ]
    for label, key in metrics:
        print(f'{label:<30}', end='')
        for name, data in results.items():
            val = data.get(key, 'N/A')
            if isinstance(val, float):
                print(f'{val:<20.2f}', end='')
            else:
                print(f'{val:<20}', end='')
        print()
    print('='*70)
```

BUILDING THE IMAGE REGISTERY

```python
import cv2
import numpy as np
import os

registry = []

for filename in os.listdir(MASTER_IMAGES):
    if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue
    img = cv2.imread(os.path.join(MASTER_IMAGES, filename))
    if img is None:
        continue
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    brightness = float(np.mean(hsv[:,:,2]))
    source = 'exdark' if 'rf.' in filename else 'visdrone'
    registry.append({'filename': filename, 'brightness': brightness, 'source': source})

print(f'Total: {len(registry)}')
print(f'ExDark: {sum(1 for r in registry if r["source"] == "exdark")}')
print(f'VisDrone: {sum(1 for r in registry if r["source"] == "visdrone")}')
```

```python
print(f'Master images: {len(os.listdir(MASTER_IMAGES))}')
print(f'ExDark source: {len(os.listdir(EXD_IMAGES))}')
print(f'VisDrone source: {len(os.listdir(VIS_IMAGES))}')
```

```python
all_brightness = [r['brightness'] for r in registry]

p20 = np.percentile(all_brightness, 20)
p40 = np.percentile(all_brightness, 40)
p60 = np.percentile(all_brightness, 60)
p80 = np.percentile(all_brightness, 80)

print(f'Unified brightness thresholds:')
print(f'  Very Dark:   0     - {p20:.1f}')
print(f'  Dark:        {p20:.1f} - {p40:.1f}')
print(f'  Moderate:    {p40:.1f} - {p60:.1f}')
print(f'  Bright:      {p60:.1f} - {p80:.1f}')
print(f'  Very Bright: {p80:.1f} - 255')

def get_bin(brightness):
    if brightness < p20:   return 0
    elif brightness < p40: return 1
    elif brightness < p60: return 2
    elif brightness < p80: return 3
    else:                  return 4

for r in registry:
    r['bin'] = get_bin(r['brightness'])

from collections import Counter
bin_counts = Counter(r['bin'] for r in registry)
print(f'\nImages per bin:')
for b, count in sorted(bin_counts.items()):
    print(f'  Bin {b}: {count} images')
```

```python
import random
random.seed(42)
np.random.seed(42)

train_files = []
val_files   = []
test_files  = []

for bin_id in range(5):
    bin_images   = [r for r in registry if r['bin'] == bin_id]
    exdark_bin   = [r for r in bin_images if r['source'] == 'exdark']
    visdrone_bin = [r for r in bin_images if r['source'] == 'visdrone']

    random.shuffle(exdark_bin)
    random.shuffle(visdrone_bin)

    def split_source(images):
        n = len(images)
        n_train = int(n * 0.70)
        n_val   = int(n * 0.15)
        train = images[:n_train]
        val   = images[n_train:n_train + n_val]
        test  = images[n_train + n_val:]
        return train, val, test

    exd_train, exd_val, exd_test = split_source(exdark_bin)
    vis_train, vis_val, vis_test = split_source(visdrone_bin)

    train_files.extend(exd_train + vis_train)
    val_files.extend(exd_val + vis_val)
    test_files.extend(exd_test + vis_test)

    print(f'Bin {bin_id}: '
          f'train={len(exd_train)+len(vis_train)} '
          f'(exd={len(exd_train)}, vis={len(vis_train)}) | '
          f'val={len(exd_val)+len(vis_val)} '
          f'(exd={len(exd_val)}, vis={len(vis_val)}) | '
          f'test={len(exd_test)+len(vis_test)} '
          f'(exd={len(exd_test)}, vis={len(vis_test)})')

print(f'\nFinal split:')
print(f'Train: {len(train_files)}')
print(f'Val:   {len(val_files)}')
print(f'Test:  {len(test_files)}')
print(f'Total: {len(train_files)+len(val_files)+len(test_files)}')
```

```python
import shutil

def copy_split(file_list, images_dir, labels_dir, split_name):
    copied = 0
    missing = 0
    for record in file_list:
        filename = record['filename']
        stem = os.path.splitext(filename)[0]
        src_img = os.path.join(MASTER_IMAGES, filename)
        src_lbl = os.path.join(MASTER_LABELS, f'{stem}.txt')
        if not os.path.exists(src_lbl):
            missing += 1
            continue
        shutil.copy2(src_img, os.path.join(images_dir, filename))
        shutil.copy2(src_lbl, os.path.join(labels_dir, f'{stem}.txt'))
        copied += 1
    print(f'{split_name}: {copied} copied, {missing} missing labels')

copy_split(train_files, TRAIN_IMAGES, TRAIN_LABELS, 'Train')
copy_split(val_files,   VAL_IMAGES,   VAL_LABELS,   'Val')
copy_split(test_files,  TEST_IMAGES,  TEST_LABELS,  'Test')
```

```python
for split_name, img_dir, lbl_dir in [
    ('Train', TRAIN_IMAGES, TRAIN_LABELS),
    ('Val',   VAL_IMAGES,   VAL_LABELS),
    ('Test',  TEST_IMAGES,  TEST_LABELS)
]:
    imgs = set(os.path.splitext(f)[0] for f in os.listdir(img_dir))
    lbls = set(os.path.splitext(f)[0] for f in os.listdir(lbl_dir))
    print(f'{split_name}: {len(imgs)} images, {len(lbls)} labels, unmatched={len(imgs-lbls)}')
```
